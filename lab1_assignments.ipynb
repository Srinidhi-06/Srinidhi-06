```python
#question 1
def count_variables_and_consonants(s):
    variables = 0
    consonants = 0
    for i in s:
        if i in ['a','e','i','o','u','A','E','I','O','U']:
            variables = variables+1
        else:
            consonants = consonants+1
    return variables,consonants
       
s = input("Enter a string to count the variables and consonants ")
variables,consonants = (count_variables_and_consonants(s))
print("Number of vowels -",variables)
print("Number of consonants - ",consonants)

#question 2
def read_matrix(name):
    print(f"Enter rows of matrix {name}. Enter each row as space-separated numbers.")
    print("Type 'done' when you are finished.")
    
    matrix = []
    while True:
        row = input("Enter row: ")
        if row.lower() == 'done':
            break
        matrix.append([int(num) for num in row.split()])
    return matrix

def multiply_matrices(A, B):
   
    rows_A = len(A)
    cols_A = len(A[0])
    rows_B = len(B)
    cols_B = len(B[0])
    
    if cols_A != rows_B:
        return None  

   
    result = [[0] * cols_B for _ in range(rows_A)]

    
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result[i][j] += A[i][k] * B[k][j]
    
    return result


def print_matrix(matrix):
    for row in matrix:
        print(" ".join(map(str, row)))

print("Matrix A:")
matrix_a = read_matrix('A')

print("Matrix B:")
matrix_b = read_matrix('B')

product = multiply_matrices(matrix_a, matrix_b)

if product is None:
    print("Error: Matrices cannot be multiplied.")
else:
    print("Product of matrix A and B:")
    print_matrix(product)

    
#question 3
def common_elements():
    l1 = []
    l2 = []
    s1 = int(input("Enter size of list 1 elements"))
    s2 = int(input('Enter size of list 2 elements'))

    for i in range(s1):
        v = int(input())
        l1.append(v)

    for i in range(s2):
        v = int(input())
        l2.append(v)
    
    l3 = []
    for i in l1:
        if i in l2:
            l3.append(i)
            
    return l3

print(common_elements())

#question 4
def transpose():
    rows = int(input('Enter number of rows'))
    columns = int(input('Enter number of columns'))
    s = []
    for i in range(rows):
        a = []
        for j in range(columns):
            v = int(input())
            a.append(v)
        s.append(a)
    print(s)

    t = [[s[i][j] for i in range(rows)]for j in range(columns)]
    for i in range(rows):
        for j in range(columns):
            if i != j:
                t[i][j]=s[j][i]
    return t

print(transpose())
```
